## **1. Contexto**

En todo negocio, se hace indispensable conocer el comportamiento de los clientes para lograr satisfacer sus necesidades y exceder sus expectativas; así, hacer crecer el negocio y la rentabilidad. Ahora, si el negocio no tiene un lugar físico, sino que tiene un lugar en la web; el proceso de poner atención a los clientes o usuarios cambia bastante.

Para esto, la tecnología permite tomar nota de diferentes atributos que describen el comportamiento de un usuario al navegar en un sitio web; como tiempos de demora en una página o posición, todo lo relacionado a donde se da click, saltos entre páginas, entre otros.

## **2. Introducción**

El reto es analizar un conjunto de datos de este tipo para responder preguntas y obtener información útil para los intereses del negocio.

En adelante, se planteará una pregunta de interés para el negocio y se intentará aproximar una respuesta a partir de un dataset.

### 2.1) Pregunta de investigación

¿Existen diferencias clave en el comportamiento de un usuario en un e-commerce que determinen si finalmente comprará algún artículo?

### 2.2) Relevancia

Realizar un análisis profundo del comportamiento de los usuarios durante su navegación en un e-commerce es fundamental para entender aquellos factores que pueden ser determinantes para concretar una venta. Esto permite optimizar la toma de decisiones orientadas a la experiencia de usuario con el fin de aumentar el flujo de clientes y finalmente las ventas.



### 2.3) Base de datos

Para realizar un análisis que pueda responder a la pregunta de investigación y entregar información valiosa, se seleccionó la base de datos [Online Shoppers Purchasing Intention Dataset](https://archive.ics.uci.edu/dataset/468/online+shoppers+purchasing+intention+dataset) del repositorio [UCI Machine Learning](https://archive.ics.uci.edu/).

El conjunto de datos consta de vectores de 17 características correspondientes a 12.330 sesiones. Se creó para que cada sesión perteneciera a un usuario diferente en un período de un año, evitando así cualquier tendencia hacia una campaña, un día especial, un perfil de usuario o un período específicos. Finalmente, el dataset describe si cada sesión culminó o no en una venta.
