### <span style="color:#2F749F;"><strong>Ejercicio 1: C치lculo de medidas de tendencia central</strong></span>

A partir de una muestra simulada de ingresos, calcula la **media, mediana y moda**, y luego a interpretarlas.

1. Usa el n칰mero base `528310` (debes reemplazarlo con los 칰ltimos 6 d칤gitos de tu documento de identificaci칩n).  
2. Con este n칰mero genera una semilla (`np.random.seed`) para que los resultados sean reproducibles.  
3. Los ingresos se simulan con una distribuci칩n normal:  
   - Media te칩rica: `(base % 10) + 2`  
   - Desviaci칩n est치ndar: `2`  
   - Tama침o de muestra: `30`  
4. Los valores negativos se reemplazan por cero (pues los ingresos no pueden ser negativos).  
5. Grafica tus resultados (ver imagen de ejemplo)

<p align="center">
    <img src="imgs\sp_2_e1.png" alt="sp_2_e1" width="900"  height="400">
</p>


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
seed = 750946
np.random.seed(seed)
data = np.random.normal((seed % 10) + 2, 2, 100).round(1)
```

```python
mean = np.mean(data)
median = np.median(data)
mode = np.bincount(data.astype(int)).argmax()
```

```python
print(f"La media es: {mean}")
print(f"La mediana es: {median}")
print(f"La moda es: {mode}")
```

```python
fig = plt.figure(figsize=(10, 5))
sns.histplot(data, bins=10,color="salmon", edgecolor="black", kde=True)
plt.title("Distribuci칩n de ingresos")
plt.xlabel("Ingresos")
plt.ylabel("Frecuencia")
plt.axvline()
plt.axvline(mean, color="blue", linestyle="--", linewidth=2, label=f"Media: {mean:.1f}")
plt.axvline(median, color="green", linestyle="-", linewidth=2, label=f"Mediana: {median}")
plt.axvline(mode, color="red", linestyle=":", linewidth=2, label=f"Moda: {mode}")
plt.legend()
plt.show()
```

# Interpretaci칩n:

Media: Las personas pertenecientes a la base de datos tienen un ingreso promedio de 8.5 (unidades monetarias / unidad de tiempo).

Mediana: Ordenando los datos ascendentemente, el dato de la mitad es 8.25.

Moda: En la base de datos hay una cantidad mayor de personas con ingreso de 9 sobre todas las otras cantidades.


### <span style="color:#2F749F;"><strong>Ejercicio 6: Chi-Cuadrado</strong></span>

En esta parte, analiza si existe **asociaci칩n estad칤stica** entre dos variables categ칩ricas simuladas a partir de tu n칰mero de c칠dula.

1. Usa los **6 칰ltimos d칤gitos** de tu c칠dula como base para generar una semilla reproducible.
2. Simula una muestra de **60 personas** con dos variables categ칩ricas:
   - `Nivel educativo`: B치sico, Medio, Superior.
   - `Acceso a internet`: S칤, No.
3. Crea una **tabla de contingencia** entre ambas variables.
4. Aplica la **prueba de Chi cuadrado de independencia**.
5. Visualiza la distribuci칩n conjunta con un mapa de calor (ver la imagen de con la informaci칩n de ejemplo).
6. Interpreta el resultado.

<p align="center">
    <img src="imgs\sp_2_e6.png" alt="sp_2_e6" width="900"  height="700">
</p>

```python
import pandas as pd
from scipy.stats import chi2_contingency

np.random.seed(seed)
nivel_educativo = np.random.choice(["B치sico", "Medio", "Superior"], size=60)
acceso_internet = np.random.choice(["S칤", "No"], size=60)

df = pd.DataFrame({"Nivel educativo": nivel_educativo, "Acceso a internet": acceso_internet})
chi, p, dof, expected = chi2_contingency(pd.crosstab(df["Nivel educativo"], df["Acceso a internet"]))
```

```python
fig, axs = plt.subplots(1,2,figsize=(20, 7))
sns.heatmap(pd.crosstab(df["Nivel educativo"], df["Acceso a internet"]), annot=True, cmap="Blues", fmt="d",ax=axs[0])
sns.countplot(data=df, x="Nivel educativo", hue="Acceso a internet", palette="Blues", ax=axs[1])
axs[0].set_title("Tabla de contingencia")
axs[1].set_title("Distribuci칩n conjunta")
plt.show()
```

```python
print(f"El valor de Chi cuadrado es: {chi:.2f}")
print(f"El valor de p es: {p:.2f}")
```

# Interpretaci칩n

Dado que p >> 0.05, no hay evidencia estad칤stica suficiente para rechazar la hip칩tesis nula de independencia, por lo que no podemos afirmar que exista una relaci칩n entre las variables.

Esto tambi칠n se puede observar en las gr치ficas. <br>
- En la tabla de contingencia podemos ver valores bastante similares en todas las celdas.
- En la tabla de distribuci칩n conjunta no se puede ver una tendencia diferente en el acceso a internet entre los niveles de educaci칩n.
