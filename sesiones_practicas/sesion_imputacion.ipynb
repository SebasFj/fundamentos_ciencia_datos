## Ejercicio 1

```python
# Importación de librerías

import numpy as np
import pandas as pd
```

```python
#Creación del dataframe

df = pd.DataFrame({
    'Nombre': ['Ana', 'Luis', 'Pedro', None, 'Marta', 'Luis', 'Sofía'],
    'Edad': [25, np.nan, 35, 29, -5, 25, None],
    'Ciudad': ['Bogotá', 'Medellín', None, 'Medellín', 'Cali', 'Bogotá', 'Cali'],
    'Ingreso': [3500, 4800, np.nan, 5200, 5100, np.nan, 4700],
    'FechaIngreso': ['2023-01-01', '2023-01-05', None, '2023-01-10', '2023-01-12', None, '2023-01-15']
})
```

```python
df.columns
```

```python
# Datos faltantes

df.isna().sum()
```

```python
#columna "Nombre"

df["Nombre"] = df["Nombre"].fillna("Sin Nombre")
```

```python
#Columna "Edad"

mean_edad = df[df["Edad"]>0]["Edad"].mean().round()
df["Edad"] = df["Edad"].apply(lambda x: mean_edad if (x<0 or pd.isna(x)) else x)
```

```python
#Columna "Ciudad"

import unicodedata

df["Ciudad"] = df["Ciudad"].fillna("Sin Ciudad")

def quitar_tildes(texto):
    return "".join((c for c in unicodedata.normalize('NFD', texto) if unicodedata.category(c) != "Mn"))

df["Ciudad"] = df["Ciudad"].apply(quitar_tildes)

df["Ciudad"] = df["Ciudad"].apply(lambda x: x.lower())
```

```python
#Columna "Ingreso"

df["Ingreso"] = df["Ingreso"].fillna(0)
```

```python
# columna "FechaIngreso"

df["FechaIngreso"] = pd.to_datetime(df["FechaIngreso"], errors="coerce")
```

## Ejercicio 2

```python
#Dataframe Base

df = pd.DataFrame({
    'ID': [101, 102, 102, 103, 104, 104, 104],
    'Nombre': ['Ana', 'Luis', 'Luis', 'Marta', 'Carlos', 'Carlos', 'Carlos'],
    'Edad': [25, 30, 30, 29, 40, 40, 41],
    'Ciudad': ['Bogotá', 'Cali', 'Cali', 'Medellín', 'Cali', 'Cali', 'Cali'],
    'FechaRegistro': ['2023-01-01', '2023-01-05', '2023-01-05', '2023-01-10',
                    '2023-01-15', '2023-01-15', '2023-01-16']
})
```

```python
df
```

### 1 - ¿Cual es el total de registros originales?

```python
df.columns
```

```python
muestras_originales = len(df["ID"].unique())
print("Registros originales: ", muestras_originales)
```

### 2 - ¿Cuáles y cuántos son los duplicados exactos?

```python
id_duplicados = df["ID"].value_counts()[df["ID"].value_counts()>1].to_dict()
for clave, valor in id_duplicados.items():
    print(f"El ID {clave} se repite {valor} veces")
print(f"Total de duplicados: {len(id_duplicados)}")
```

### 3 - ¿Cuáles y cuántos son los duplicados por varias columnas?

```python
for col in df.columns:
    duplicados = df[col].value_counts()[df[col].value_counts()>1].to_dict()
    print(f"Columna: {col}")
    for clave, valor in duplicados.items():
        pre = "El" if col == "ID" or col == "Nombre" else "La"
        print(f"{pre} {col} {clave} se repite {valor} veces")
    print("")
```

### 4 - ¿Cuántos registros debes eliminar?

```python
muestras_eliminar = sum(id_duplicados.values()) - len(id_duplicados)
print(f"Por ID se deben eliminar {muestras_eliminar} muestras.")
```

### 5 - ¿Cuántos registros quedan después de la limpieza?

```python
print(f"Quedan un total de {df.shape[0] - muestras_eliminar} muestras.")
```
